### Looping Statement
  For

  While

#### For

```python
a = [1,3,4,5,2,100]
b = 0
for i in a:
  print(i)
  b +=i

print("final b is ", b)
```

```python
a = [1,3,4,5,2,100]
b = 0

for i in a:
  print(i)
  b = b + i
else:
  print("No elements found, so final sum of b is ", b)
```

#### While

```python
a = 10
b = 1
while a != b:
  print("a is not met with b", b)
  b = b + 1
else:
  print("b is equal to a", b)
```
